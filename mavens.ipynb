# Concatinating all the Data Frame using .concat() 

```python
call_service = pd.concat([call_service2016, call_service2017, call_service2018, call_service2019, call_service2020])
pd.set_option('display.max_columns', None)
call_service.head()
```

# Calculating the mean(Average) of the time of response using the time a dispatch was made to the 

```python
call_service['TimeDispatch'] = pd.to_datetime(call_service['TimeDispatch'], infer_datetime_format=True)

# now for your answer, get the average of the timestamp column
print(call_service['TimeDispatch'].mean())
```

```python
call_service['TimeArrive'] = pd.to_datetime(call_service['TimeArrive'], infer_datetime_format=True)

# now for your answer, get the average of the timestamp column
print(call_service['TimeArrive'].mean())
```

```python
Mean_Aggregate = (call_service['TimeDispatch'].mean() - call_service['TimeArrive'].mean())/2
Mean_Aggregate
```
